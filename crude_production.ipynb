```python
# Matthew McCarroll created 8/31/2020
import numpy as np
import pandas as pd
```

```python
# opens the downloaded file from https://www.eia.gov/dnav/pet/pet_crd_crpdn_adc_mbblpd_m.htm
xls = pd.ExcelFile('PET_CRD_CRPDN_ADC_MBBLPD_M.xls')
data = pd.read_excel(xls, 'Data 1', skiprows=2)
```

```python
data.tail()
```

```python
data.iloc[-1]
```

```python
# I only want data from 2000 and on up
data=data.loc[data['Date']>='2000']
```

```python
data.mean()
```

```python
data.head()
# remove unit measurement from headers
actual =[string.replace('Field Production of Crude Oil (Thousand Barrels per Day)','').rstrip() for string in data.columns]
data.columns=actual
actual =[string.replace('Crude Oil Production (Thousand Barrels per Day)','').rstrip() for string in data.columns]
data.columns=actual
```

```python
# add states that do not produce oil for mapping purposes
data['New Jersey'] = 0
data['Georgia'] = 0
data['South Carolina'] = 0
data['North Carolina'] = 0
data['Delaware'] = 0
data['Maryland'] =0
data['Vermont']=0
data['New Hampshire']=0
data['Rhode Island']=0
data['Massachusetts']=0
data['Connecticut']=0
data['Maine']=0
data['Washington']=0
data['Oregon']=0
data['Wisconsin'] = 0
data['Minnesota']=0
data['Iowa']=0
```

```python
data.info()
```

```python
data.tail(10)
```

```python
# transform data to assign PADD region and state location to each value for each date
production=pd.DataFrame()
for col in data.columns.drop(['Date']):
    section=pd.DataFrame()
    section['date']=data['Date']
    section['location']=col
    section['production']=data[col]
    production=production.append(section)
```
